```python
#Part 1
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')
```

```python

# Load the data from the logins.json file
data = pd.read_json('logins.json')

# Display the first few rows to inspect the data
print(data.head())

print(data.info())
```

```python
## grouping by 15 mins
data = data.reset_index()
data.drop('index', axis = 1, inplace = True)
data = data.set_index('login_time')
data['count'] = 1
data = data.resample('15T').sum()
```

```python
print(data.head())
```

```python
#Plotting the data series

plt.figure(figsize = (18,4))
plt.plot(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Resample for daily data

# Resample the data to daily intervals and calculate the mean for each day
daily_data = data.resample('D').mean()

# Plot the daily data
plt.figure(figsize=(12, 6))
daily_data['count'].plot(title='Daily Mean Login Counts', xlabel='Date', ylabel='Login Counts')
plt.grid(True)
plt.show()

```

```python
#looks like there is some seasonality in this data series! Let's plot the PACF and ACF

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


# Plot autocorrelation and partial autocorrelation plots
plot_acf(daily_data['count'])
plot_pacf(daily_data['count'])

```

```python
#The ACF plot shows significant values at lag 7, 14 and 21. Which are clearly weekly intervals. 
#PACF Plot shows significant values at lag 2, 3, 6,7, as well. 
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

#S = 7 for seasonal effects

p,d,q = 2,0,1
S = 7
P,D,Q = 0,0,0
# Define the SARIMA model with appropriate order parameters
sarima_model = SARIMAX(data['count'], order=(p, d, q), seasonal_order=(P, D, Q, S))

# Fit the SARIMA model to your data
sarima_results = sarima_model.fit()

# Print a summary of the SARIMA model
print(sarima_results.summary())
```

```python
# Get the residuals from the SARIMA model
residuals = sarima_results.resid

# Plot the residuals
plt.figure(figsize=(12, 6))
plt.plot(residuals)
plt.title('Residual Plot')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.grid(True)
plt.show()

#Residuals look nonstationary
```

```python
from statsmodels.graphics.tsaplots import plot_acf

# Plot the ACF of the residuals
plot_acf(residuals)

```
